```python
url = 'https://raw.githubusercontent.com/ybifoundation/Dataset/main/Movies%20Recommendation.csv'
df = pd.read_csv(url,encoding='unicode_escape')
```

```python
df_features = df[['Movie_Genre','Movie_Spoken_Language','Movie_Keywords','Movie_Production_Country','Movie_Cast']]
```

```python
df_features.info()
```

```python
df_features = df_features.dropna()
```

```python
X = df_features['Movie_Genre']+' '+df_features['Movie_Spoken_Language']+' '+df_features['Movie_Keywords']+' '+df_features['Movie_Production_Country']+' '+df_features['Movie_Cast']
```

```python
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python

X = tfidf.fit_transform(X)
```

```python
X
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(X)
```

```python
print(Similarity_Score)
Similarity_Score.shape
```

```python
Fav_Movie_Name = input('Enter your favourite movie name:')
```

```python
All_Movies_Title = df['Movie_Title'].tolist()
```

```python
import difflib as dl
```

```python
Movie_Recommendation =  dl.get_close_matches(Fav_Movie_Name,All_Movies_Title)
```

```python
print(Movie_Recommendation)
Close_Match = Movie_Recommendation[0]
```

```python
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
print(Recommendation_Score)
len(Recommendation_Score)
```

```python
Sorted_Similer_Movies = sorted(Recommendation_Score, key=lambda x:x[1],reverse = True)
print(Sorted_Similer_Movies)
```
